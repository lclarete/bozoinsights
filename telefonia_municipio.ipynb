```python
# import a ods parser
from pandas_ods_reader import read_ods

# set the file path
ods_path = 'Municípios atendidos com SMP e tecnologias.ods'

# define the column's names
df_columns = ['IBGE',
             'UF', 
             'AR',
             'Municipio',
             'algar_2G',
             'algar_3G',
             'algar_4G',
              'claro_2G',
              'claro_3G',
              'claro_4G',
              'nextel_2G',
              'nextel_3G',
              'nextel_4G',              
              'oi_2G',
              'oi_3G',
              'oi_4G',              
              'tim_2G',
              'tim_3G',
              'tim_4G',
              'vivo_2G',
              'vivo_3G',
              'vivo_4G',              
              'sercomtel_2G',
              'sercomtel_3G',
              'sercomtel_4G',
             ]

# read dataframe
df = read_ods(ods_path, sheet=1,
              # set the column's names
             columns = df_columns,
             headers=False)

# delete unuseful rows
df = df.drop([0,1], axis=0)

# reset index
df = df.reset_index()

# drop unuseful columns
df = df.drop('index', axis=1)
```

```python
df = df.replace('None', False).replace('NÃO', False).replace('SIM', True)
```
