# Functions

```python
import pandas as pd

# identify each city is not cover by 2G, 3G or 4G network
def net_xG(net_type, df=df):
    IBGE = df['IBGE']
    net_cols = df.columns[df.columns.str.contains(net_type)]
    df_net_type = df[net_cols].sum(axis=1)
    df_net = pd.concat([IBGE, df_net_type], axis=1)
    df_net.columns = ['IBGE', net_type]
    df_net[net_type] = [True if x > 0 else False for x in df_net[net_type]]
    return(df_net)
```

```python
# define networks
net_types = ['2G', '3G', '4G']

# call net_xG function on each network
list_types = [net_xG(i) for i in net_types]

# concat the list
df_types = pd.concat(list_types, axis=1)

# drop unuseful columns
df_types = df_types.drop('IBGE', axis=1)

# concat with df
df_total = pd.concat([df, df_types], axis=1)

# select columns
df_xg = df_total[['IBGE', 'UF', 'Municipio', '2G', '3G', '4G']]
```

```python
xg = {True:1, False:0}
xg_cols = ['2G', '3G', '4G']
```

```python
for i in xg_cols:
    df_xg[i] = df_xg[i].map(xg)
```

```python
df_xg_total = df_xg.pivot_table(
    index='Municipio',
    values= ['2G', '3G', '4G'],
    aggfunc='sum'
)
```

```python
df_xg_total['xg_total'] = df_xg_total.sum(axis=1)
```
