```python
# import a ods parser
from pandas_ods_reader import read_ods

# set the file path
ods_path = 'Municípios atendidos com SMP e tecnologias.ods'

# define the column's names
df_columns = ['IBGE',
             'UF', 
             'AR',
             'Municipio',
             'algar_2G',
             'algar_3G',
             'algar_4G',
              'claro_2G',
              'claro_3G',
              'claro_4G',
              'nextel_2G',
              'nextel_3G',
              'nextel_4G',              
              'oi_2G',
              'oi_3G',
              'oi_4G',              
              'tim_2G',
              'tim_3G',
              'tim_4G',
              'vivo_2G',
              'vivo_3G',
              'vivo_4G',              
              'sercomtel_2G',
              'sercomtel_3G',
              'sercomtel_4G',
             ]

# read dataframe
df = read_ods(ods_path, sheet=1,
              # set the column's names
             columns = df_columns,
             headers=False)

# delete unuseful rows
df = df.drop([0,1], axis=0)

# reset index
df = df.reset_index()

# drop unuseful columns
df = df.drop('index', axis=1)

# transform network cols into bool type
df = df.replace('NÃO', False).replace('SIM', True)
```

```python
import pandas as pd

# identify each city is not cover by 2G, 3G or 4G network
def net_xG(net_type, df=df):
    IBGE = df['IBGE']
    net_cols = df.columns[df.columns.str.contains(net_type)]
    df_net_type = df[net_cols].sum(axis=1)
    df_net = pd.concat([IBGE, df_net_type], axis=1)
    df_net.columns = ['IBGE', net_type]
    df_net[net_type] = [True if x > 0 else False for x in df_net[net_type]]
    return(df_net)
```

```python
# define networks
net_types = ['2G', '3G', '4G']

# call net_xG function on each network
list_types = [net_xG(i) for i in net_types]

# concat the list
df_types = pd.concat(list_types, axis=1)

# drop unuseful columns
df_types = df_types.drop('IBGE', axis=1)

# concat with df
df_total = pd.concat([df, df_types], axis=1)

# select columns
df_networks = df_total[['IBGE', 'UF', 'Municipio', '2G', '3G', '4G']]
```
